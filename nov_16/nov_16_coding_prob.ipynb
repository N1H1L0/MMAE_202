```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz 

points = pd.read_excel("sampled_points(1).xlsx", engine = "openpyxl")

x = points["x"].values
y = points["y"].values



ab_stress = []
for i in range(500):
  ab_stress.append(100/(np.pi* y[i]**2))

bc_stress = []
for i in range(500, 1000):
  bc_stress.append(50/(np.pi*y[i]**2))


normal_stress = np.concatenate([ab_stress, bc_stress])


plt.scatter(x, normal_stress)
```

```python

d1 = trapz( 100/ (70 * np.pi*y[:500]**2), dx = 0.001001)
d2 = trapz( 50/ (70 * np.pi*y[500:]**2), dx = 0.001001)

print("Total deformation in gigapascals is", d1+ d2)
```
