```python
#First import your libraries
import numpy as np
import matplotlib.pyplot as plt


F = 5

#Create your crankshaft function
def crankshaft(theta,F):
  #Assume values of l and r
  r=0.3
  l=1
  #Do further calculations to find out torque T in terms of F and theta
  alpha = r*np.arcsin(r*(np.sin(theta)/l))
  beta = 180-(r*np.arcsin(r*(np.sin(theta)/l)))-theta
  s=np.sqrt((F*np.sin(alpha))**2+(F*np.cos(alpha))**2)
  Fr=np.sqrt((s*np.sin(theta))**2+(s*np.cos(theta))**2)
  Ft=Fr*np.tan(90-theta)

  #return the torque
  return Ft*r


#Create a vector for theta where you store all values of theta over which you will do the crankshaft computation
#Use the linspace function of numpy
theta = np.linspace(0, 2*np.pi, 360, True)
#Create a vector for T which is the same length as theta and initialize with 0 values
#Use the zeros_like function of numpy

#Run a loop where you calculate the torque values at the theta values
crankshaft(theta, F)
T=[]

np.zeros_like(T)

for i in range(0, 360):
  T.append(crankshaft(i,F))

#Plot
plt.plot(theta,T)
plt.show()
```
