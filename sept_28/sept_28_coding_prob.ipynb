```python
import numpy as np
import matplotlib.pyplot as plt

def rod_b(t):
    p = 100
    return p/np.sin(t * (np.pi/180))

def rod_cy(t):
    p = 100
    return p*(1-(1/np.tan(t* (np.pi/180))))

rod_cx = 100
Cx = []
B=[]
Cy = []

for i in range(1, 90):
  Cx.append(rod_cx)
  B.append(rod_b(i))
  Cy.append(rod_cy(i))

plt.plot(Cx, label = "Cx", color = "r", marker = "o")
plt.plot(B, label = "B", color="g", marker = "x")
plt.plot(Cy, label = "Cy", color = "b", marker = "s")

#showing the legend 
plt.legend()

#displaying the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
